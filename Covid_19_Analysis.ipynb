## 4. Filter the dataframe with only this columns
['continent','location','date','total_cases','total_deaths','gdp_per_ca
pita','
human_development_index'] and update the data frame.

```python
df_new = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

```python
df_new
```

## 5. Data Cleaning

#### a. Remove all duplicates observations


```python
df_new.duplicated().value_counts()   # DataFrame.duplicated(subset = None, keep = 'first')
```

```python
df_new.drop_duplicates() # use for remove duplicates
```

#### b. Find missing values in all columns

```python
df_new.isnull() # we can use isna
```

#### c. Remove all observations where continent column value is missing 

```python
df_new2 = df_new.dropna(axis=0, subset=['continent'])
```

```python
df_new2
```

#### d. Fill all missing values with 0

```python
df_new2.fillna(0) #[df['total_cases'].mean()]
```

## 6. Date time format :

 ### a. Convert date column in datetime format using
pandas.to_datetime

```python
df_new2['date'] = pd.to_datetime(df_new2['date'])
```

```python
df_new2
```

```python
df_new2.dtypes
```

## b. Create new column month after extracting month data from
date
 column.

```python
df_new2['month'] = df_new2['date'].dt.month
```

```python
df_new2
```

## 7. Data Aggregation:


## a. Find max value in all columns using groupby function on
'continent'

```python
df_new2.groupby('continent').max()
```

```python
df_new2.reset_index()
```

### b. Store the result in a new dataframe named 'df_groupby'.
### (Use df_groupby dataframe for all further analysis)


```python
df_groupby = df_new2
```

```python
df_groupby
```

## 8. Feature Engineering :

### a. Create a new feature 'total_deaths_to_total_cases' by ratio of
### 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
df_groupby
```

## 9. Data Visualization :

### a. Perform Univariate analysis on 'gdp_per_capita' column by
### plotting histogram using seaborn dist plot.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,7))
plt.title('GDP-Density graph')
sns.distplot(df_groupby['gdp_per_capita'],kde=True,hist=True)
```

### b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'


```python
sns.jointplot(data=df_groupby,x='total_cases',y ='gdp_per_capita',kind='scatter')
```

#### c. Plot Pairplot on df_groupby dataset.

```python
sns.pairplot(df_groupby)
```

```python
df_groupby.corr() # for basic understanding of correlation 
```

#### d. Plot a bar plot of 'continent' column with 'total_cases' .
#### Tip : using kind='bar' in seaborn catplot


```python
plt.figure(figsize=(10,7))
sns.catplot(data=df_groupby,x='continent',y='total_cases',kind='bar')
```
